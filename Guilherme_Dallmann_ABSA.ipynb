````python
try:
    import cPickle as pickle
except:
    import pickle
import gzip
````

````python
def save(object, filename, protocol = -1):
  """Save an object to a compressed disk file.
      Works well with huge objects.
  """
  file = gzip.GzipFile(filename, 'wb')
  pickle.dump(object, file, protocol)
  file.close()

def load(filename):
  """Loads a compressed object from disk
  """
  file = gzip.GzipFile(filename, 'rb')
  object = pickle.load(file)
  file.close()

  return object
````

````python
dic_ANEW_Br = {}

# Iterate using .iterrows() to avoid potential index issues
for index, row in df_final_01.iterrows():
  palavra = str(row['Palavra'])
  if palavra not in dic_ANEW_Br:
    dic_ANEW_Br[palavra] = {'polaridade': []}
````

````python
for index, row in df_final_01.iterrows():
  palavra = str(row['Palavra'])  # Get the word at index i
  if float(row['Valência_Média']) > 6:
    dic_ANEW_Br[palavra]['polaridade'].append(1)  # Append to the 'polaridade' list for the word
  elif float(row['Valência_Média']) > 4 and float(row['Valência_Média']) < 4 :
    dic_ANEW_Br[palavra]['polaridade'].append(0)
  else:
    dic_ANEW_Br[palavra]['polaridade'].append(-1)
````

````python
def moda(dicionario):
    modas = {}
    for item, valores in dicionario.items():
        polaridades = valores['polaridade']
        try:
            modas[item] = mode(polaridades)
        except StatisticsError:
            modas[item] = "Moda indefinida"
    return modas
````

````python
moda_por_item = moda(dic_ANEW_Br)

for item, moda in moda_por_item.items():
    print(f"A moda da polaridade para '{item}' é: {moda}")
````
